```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

spotify_df = pd.read_csv("../data/Cleaned_Spotify_2024_Global_Streaming_Data.csv")
spotify_df.head()
```

```python
#group by genre and calculate the average total streams
avg_streams = spotify_df.groupby('Genre')['Total Streams (Millions)'].mean().sort_values(ascending=False)
avg_streams
```

```python
plt.bar(avg_streams.index, avg_streams.values)
plt.show()
```

```python
plt.bar(avg_streams.index, avg_streams.values)

plt.xlabel('Genre')
plt.ylabel('Average Total Streams (Millions)')
plt.title('Average Total Streams by Genre')

plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(avg_streams.index, avg_streams.values)

plt.xlabel('Genre')
plt.ylabel('Average Total Streams (Millions)')
plt.title('Average Total Streams by Genre')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(avg_streams.index, avg_streams.values)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xlabel('Genre')
plt.ylabel('Average Total Streams (Millions)')
plt.title('Average Total Streams by Genre')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()

plt.show()
```

```python
avg_artist_streams = spotify_df.groupby('Artist')['Total Streams (Millions)'].mean().sort_values(ascending=False)
avg_artist_streams
```

```python
plt.figure(figsize=(10,6))
colors = ['#BE5E7B']*len(avg_artist_streams)

max_index = avg_artist_streams.idxmax()
highlight_position = avg_artist_streams.index.get_loc(max_index)
colors[highlight_position] = "#000102"

plt.bar(avg_artist_streams.index, avg_artist_streams.values, color = colors, width=0.8)
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xlabel('Artist')
plt.ylabel('Average Total Streams (Millions)')
plt.title('Average Total Streams by Artist')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(10,6))

special_artist = 'Ariana Grande'
colors =["#000102" if artist == special_artist else 'grey' for artist in avg_artist_streams.index]

plt.bar(avg_artist_streams.index, avg_artist_streams.values, color = colors, width=0.8)
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xlabel('Artist')
plt.ylabel('Average Total Streams (Millions)')
plt.title('Ariana Grande Ranks 2nd in Total Average Streams')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.scatter(
    spotify_df['Streams Last 30 Days (Millions)'], 
    spotify_df['Total Streams (Millions)'],
    color = '#08cc15',
    alpha = 0.8,
    s = 25)

plt.xlabel('Streams this Month')
plt.ylabel('Total Streams')
plt.title('No Apparent Correlation Between Monthly Listens and Total Streams')

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()

```

```python
plt.figure(figsize=(10,6))

featured_genre = 'Pop'

colors =['#08cc15' if genre == featured_genre else "#a6bba8" for genre in spotify_df['Genre']] 

plt.scatter(
    spotify_df['Streams Last 30 Days (Millions)'], 
    spotify_df['Total Streams (Millions)'],
    color = colors,
    alpha = 0.8,
    s = 25)

plt.xlabel('Streams this Month')
plt.ylabel('Total Streams')

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_title('Do Pop Songs Achieve Higher Monthly Listens or Total Streams?', fontsize=18, pad=25)

for spine in ax.spines.values():
    spine.set_linewidth(0.25)
    spine.set_alpha(0.5)
    
plt.tight_layout()
plt.show()
```

```python
yearly_avg = spotify_df.groupby('Release Year')['Total Streams (Millions)'].mean().round(2)
yearly_avg
```

```python
plt.figure(figsize=(12,6))
plt.plot(
    yearly_avg.index, 
    yearly_avg.values,
    marker='+',
    linestyle='--',
    color = '#20a32b',
    linewidth=0.5)
plt.xlabel('Release Year')
plt.ylabel('Average Total Streams')
plt.title('Did Listeners Stream Less in 2022?', fontsize= 18, pad= 20)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.axvline(
    2022,
    color= '#063b0a', 
    linewidth=0.5,
    alpha=0.7)

ax.text(
    2022+0.05,
    2225,
    'Dip in Total Streams',
    color='#063b0a',
    fontsize=8
)

plt.ylim(0, 3500)

plt.tight_layout
plt.show()
```
